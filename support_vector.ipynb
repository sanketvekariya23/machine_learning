```python
new = pd.get_dummies(df.species)
```

```python
new
```

```python
final = pd.concat([df , new],axis = "columns")
```

```python
final
```

```python
final1 = final.drop(['species','Iris-virginica'] , axis = 'columns')
```

```python
final1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = final1.drop(['Iris-setosa','Iris-versicolor'], axis = 'columns')
```

```python
y = final1['Iris-versicolor']
```

```python
x_train, x_test, y_train , y_test = train_test_split(x,y , test_size = 0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train , y_train)
```

```python
model.score(x_test , y_test)
```
